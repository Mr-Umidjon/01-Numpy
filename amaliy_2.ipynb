### 4.	Null qiymatlarni yoki fillna() funksiyasi orqali to‘ldirish yoki dropna() funksiyasi orqali tashlab yuborish

```python
automobile_data.head(10)
```

```python
automobile_data_result = automobile_data.fillna(value=automobile_data.median(axis=0, numeric_only=True), axis=0, inplace=False)
```

```python
automobile_data_result['engine-type'] = automobile_data_result['engine-type'].fillna(automobile_data_result['engine-type'].mode()[0])
```
