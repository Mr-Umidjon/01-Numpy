```python
d = []
for i in range(len(df["x1"])):
  d.append(((x1 - df.iat[i, 0])**2 + (x2 - df.iat[i, 1])**2)**(1/2))
d
```

```python
df_with_dictance = df.copy()
df_with_dictance["distance"] = d
```

```python
sorted_df = df_with_dictance.sort_values("distance",  axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
sorted_df
```

```python
k = 5
knn_df = sorted_df.head(k);
knn_df
```
